```python
from matplotlib import style as st

st.use('ggplot')

x1 = [4,6,9]
y1 = [2,9,6]
x = [5,6,2]
y = [2,5,9]

plt.plot(x,y, label='first', linewidth=4)
plt.plot(x1,y1, label='second')
plt.legend()

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Ishan's Data")

plt.show()
```
