```python
#Python Program to display all prime numbers within the interval
num1 = 3
num2 = 18

for i in range(num1, num2+1):
  if(i>1):
    flag=False
    for j in range(2, i):
      if(i%j==0):
        flag=True
    if not flag:
      print(i)
```
