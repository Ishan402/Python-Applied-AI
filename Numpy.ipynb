```python
#Take sin, cos and tan of every value
val = np.array([0, 15, 30, 45, 60, 75,90])
val = np.sin(np.radians(val))
print(val)
```

```python
val1 = np.array([0, 15, 30, 45, 60, 75, 90])
val1 = np.cos(np.radians(val1))
print(np.around(val1,3))
```

```python
#Exception - tan(90)=infinity
val3 = np.array([0, 15, 30, 45, 60, 75])
val3 = np.tan(np.radians(val3))
print(np.around(val3,4))
```

##Reorganizing arrays

```python
before = np.array([[1,2,3,4], [5,6,7,8]])
print(before)

after = before.reshape([4,2])
print(after)
```

```python
#Vertically stacking Vectors and matrices
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v2,v2])
#vstack = Vertical stack
```

```python
#Horizontal Stacking Vectors and Matrices
h1 = np.ones([2,4])
h2 = np.zeros([2,2])
print(h1)
print ('------------------------------------------')
print(h2)
print('------------------------------------------')
np.hstack([h1, h2, h2])
#hstack = horizontal stack
```
