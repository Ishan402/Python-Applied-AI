**ARGUMENTS**

```python
#Arguments
def funargs(normal, *args, **kwargs):
  print(normal)
  
  for i in args:
    print(i)
  print("Now I would like to introduce some of our heroes")

  for key, value in kwargs.items():
    print(f"{key} for {value}")   #fstrings

name = ['Ishan', 'Harry', 'Srini', 'Samantha', 'Juliete']
normal = "this is normal print line"
kw = {"a":'Apple', "b":"Ball", "c":"Cat", "d":"Dog", "e":"Elephant", "f":"Fox", "g":"Goat"}
funargs(normal, *name, **kw)

#IMP NOTE - Always keep normal argument at first index and then *args at last 
#index else will give type error'''
```

**RECURSION**

```python
# Print factorial of number using iteration
def factorial(n):
  fact = 1
  for i in range(n):
    fact = fact*(i+1)
  return fact

v = factorial(5)
print(v)
```

```python
# Print factorial of number using recursion function
# n! = n * (n-1)

def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)

v = factorial(10)
print(v)
```

```python
# Print fibonacci series using recursive function
# eg - 0, 1, 1,2,3,5,8,13,21,.......

def fabonacci(n):
  if n==1:
    return 0
  elif n==2:
    return 1
  else:
    return fabonacci(n-1) + fabonacci(n-2)

v=fabonacci(8)
print(v)
```
